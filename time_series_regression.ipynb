```python
import pandas as pd
```

```python
data = pd.read_csv('data\energy_dataset.csv')
data.head()
```

forecast wind onshore day ahead is all null

```python
data["forecast wind offshore eday ahead"].drop_duplicates().head()
```

```python
data = data.drop(columns=["generation hydro pumped storage aggregated","forecast wind offshore eday ahead"])
data.head()
```

```python
data["time"] = pd.to_datetime(data["time"],utc=True)
data = data.sort_values("time")
data.head()
```

```python
data["Lag1_price_actual"] = data["price actual"].shift(1)
data["Lag2_price_actual"] = data["price actual"].shift(2)
data["Lag3_price_actual"] = data["price actual"].shift(3)
data["Lag4_price_actual"] = data["price actual"].shift(4)
data["Lag5_price_actual"] = data["price actual"].shift(5)
data["Lag6_price_actual"] = data["price actual"].shift(6)
data.head()
```

```python
data.transpose().head(100)
```

```python
X = data.drop(columns=["time","price actual"]).fillna(0)
y = data[["price actual"]].fillna(0)

split_point = int(len(data)*0.8)
X_train,X_test = X[:split_point], X[split_point:]
y_train,y_test = y[:split_point], y[split_point:]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
```

```python
import numpy as np

mse = np.mean((y_test - predictions)**2)
mae = np.mean(np.abs(y_test - predictions))
mape = np.mean(np.abs(y_test - predictions)/y_test)*100
print("MSE is:",mse)
print("MAPE is:",mape)
print("MAE is:",mae)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(12,6))
plt.plot(y_test.index, y_test, label='Actual', color = 'green')
plt.plot(y_test.index, predictions, label='Actual', color = 'black')
plt.title("Time Series Regression Price Actual")
plt.xlabel("Time")
plt.ylabel('Price Actual')
plt.legend()
plt.tight_layout()
plt.show()
```

Regression with Normalizing

```python
cols = X.columns
X_norm = pd.DataFrame()

for col in cols:
    if np.max(X[col]) >0:
        X_norm[col] = X[col] / np.max(X[col])


split_point = int(len(data)*0.8)
X_train,X_test = X_norm[:split_point], X_norm[split_point:]
y_train,y_test = y[:split_point], y[split_point:]
```

```python

from sklearn.linear_model import LinearRegression

model_2 = LinearRegression()
model_2.fit(X_train,y_train)
predictions = model_2.predict(X_test)
```

```python
import numpy as np

mse_2 = np.mean((y_test - predictions)**2)
mae_2 = np.mean(np.abs(y_test - predictions))
mape_2 = np.mean(np.abs(y_test - predictions)/y_test)*100
print("MSE is:",mse,"MSE_2 is:",mse_2)
print("MAPE is:",mape,"MAPE_2 is:",mape_2)
print("MAE is:",mae,"MAE is:",mae_2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(12,6))
plt.plot(y_test.index, y_test, label='Actual', color = 'green')
plt.plot(y_test.index, predictions, label='Actual', color = 'black')
plt.title("Time Series Regression Price Actual")
plt.xlabel("Time")
plt.ylabel('Price Actual')
plt.legend()
plt.tight_layout()
plt.show()
```
